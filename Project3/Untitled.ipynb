```python
def is_prime(n):
    if n <= 1:
        return False
    sqrt_n = int(np.sqrt(n)) + 1
    for i in range(2, sqrt_n):
        if n % i == 0:
            return False
    return True

def mu(n):
    if n <= 1:
        return 1
    num_p = 0
    sqrt_n = int(np.sqrt(n)) + 1
    for i in range(2, sqrt_n):
        if is_prime(i):
            if n % (i * i) == 0:
                return 0
            if n % i == 0:
                num_p += 1
    if num_p % 2 == 0:
        return -1
    return 1

def riemann_f(x):
    first = li(x)
    second = 0
    for i in range(1, 40):
        second += 2 * ei(log(x) * zetazero(i)).real
    third = -log(2)
    fourth = 0.140010101143
    return first - second + third + fourth

def pi(x):
    n = 1
    result = 0
    while x ** (1/n) >= 2:
        if mu(n) == 0:
            n += 1
            continue
        term = (riemann_f(x ** (1/n))).real
        result += mu(n) / n * term
        print(n, result.real)
        n += 1
    return int(result.real)

pi(100000000)
```
